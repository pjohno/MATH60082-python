5. You are given that
$$
\phi(\eta,\xi) = e^{-\lambda_b\eta} e^{-\lambda_s\xi} 
$$
and
$$
\Gamma(x,t) = (1-\phi(x^2t,\sqrt{t})) + \sin(\phi (t^{-1},x t^{-2})).
$$
Write two functions to evaluate these quantities. Calculate $\Gamma$ when $\lambda_b = 1.3$, $\lambda_s =0.26$, $x= \pi^2$ and $t=\sqrt{3}$. 
	
-	Think about how you can include parameters in your functions.

```python
# we need exp function and sin function
from math import exp,sin,sqrt
# create a function for phi, although both eta/xi
# are variables, we need the parameters as well!
def phi_func( eta , xi , lambda_b, lambda_s):
    return exp( -lambda_b*eta)*exp( -lambda_s*xi)
```

```python
# first check phi function working
print( phi_func( 0.1 , 0.3, 0.2 , 0.5))
```

```python
# now write Gamma func, again implicitly we need "lambda_b" and 
# "lambda_s" to be known
def Gamma_func(x,t,lambda_b,lambda_s):
    return ( 1.0 - phi_func( x*x*t, sqrt(t),lambda_b,lambda_s)) - sin( phi_func( 1/t, x/t/t,lambda_b,lambda_s ) )
```

```python
print( Gamma_func( 1.2 , 0.6, 0.2 , 0.5))
```

```python
def swapValues(a,b):
    aTemp = numpy.copy(a)
    a = b
    b = aTemp
    return a,b
```
