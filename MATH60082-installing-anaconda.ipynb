# Installing Anaconda on Windows

This guide gives the installation process for Windows, but the process on other operating systems is very similar. Installer screens and the download page change over time, so follow the closest matching options. Screenshots may look slightly different from the current UI. Once Anaconda is installed, the way things work inside it, and for example VS Code (Spyder or any other app launched inside), is almost the same.

- First go to Anaconda website and download the latest version for your system (choose the 64-bit Windows installer).
- If you are using Anaconda for commercial or enterprise work, check the current license terms on the download page.

[https://www.anaconda.com/download](https://www.anaconda.com/download)

## macOS note

- Choose the macOS installer that matches your hardware (Apple Silicon or Intel).
- If `conda` is not found in Terminal, run `conda init zsh` (or `conda init bash`) and restart Terminal.
- If VS Code does not appear in Navigator, go to the VS Code website to download and install it, then install the Python extension and select your Anaconda interpreter.
