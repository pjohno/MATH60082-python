4. A student is asked to write a code where the user is prompted to input the current year, and it then calculates how many years since Manchester United last won the Premier League. Fix any errors in the student's code which can be found [here (click to download)](https://raw.githubusercontent.com/pjohno/MATH60082-python/main/examples/MATH60082-support-example-3.py).

```python
#!/usr/bin/env python3
""" 
A program to prompt user for current date, and then calculate how many years it is since Manchester United last won the league
"""
# libraries and methods
from datetime import timedelta,date

# declare variables and assign values
dateLastWon = date(2013,5,19)               # date on last day of 2012/2013 season was 19th May 2013
day = int(input("Input day of month (1-31)"))    # input day of month
month = int(input("Input month (1-12)"))         # input month
year = int(input("Input year "))                 # input year
dateCurrent = date(year,month,day)          # create current date

# calculate the number of years
yearsSinceWon = ( (dateCurrent-dateLastWon).days ) / 365.0

# output the results
print("On ",dateCurrent,", it will be ",yearsSinceWon," years since Manchester united last won the Premier League.")
```
